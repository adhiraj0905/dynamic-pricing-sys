```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
try:
    data = pd.read_csv("../ride_sharing_data.csv")
except FileNotFoundError:
    print("Data file not found. Please ensure the CSV file is in the correct directory.")

print("Data Head:")
print(data.head())
```

```python
print("\nData info:")
data.info()
```

```python
print("\nMissing values:")
print(data.isnull().sum())
```

```python
print("Data Description:")
print(data.describe())  
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=30, kde=True)
plt.title('Distribution of Ride Prices')
plt.show()
```

```python
categorical_features = ['Time_of_Day','Weather']
for feature in categorical_features:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=feature, y="Price", data=data)
    plt.title(f'Price by {feature}')
   
    plt.show()
```

```python
numerical_features = ['Distance','Demand','Base_Price','Weather_Multiplier']
```

```python
for feature in numerical_features:
    plt.figure(figsize=(10,6))
    sns.scatterplot(x=feature,y="Price", data=data, alpha=0.5)
    plt.title(f'Price vs {feature}')
    plt.show()
```

```python
numerical_data=data.select_dtypes(include=[np.number])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(numerical_data.corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correalation Heatmap(numerical features)")
plt.show()
```
